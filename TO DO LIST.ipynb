#                                 TO DO LIST

```python
def menu():
    print("*************** MENU ***********************")
    print("enter '1' to add a task in the to do list")
    print("enter '2' to delete an item from the to do list")
    print("enter '3' to delete all items from the list")
    print("enter '4' to view the tasks in the list")
    print("enter '5' to exit the programme")
    
  
def user_input(your_input):
    global var
    if your_input == '1':
         add()
    elif your_input == '2':
        delete()
    elif your_input == '3':
        del_all()
    elif your_input == '4':
        viewtasks()
    elif your_input == '5':
        print("you have exit the programme")
        var = False
    else:
        print("Invalid input")

        
        
tasks_to_do = []
def add():
    print("enter s to stop adding the tasks\n")
    while True:
        tasks= input("Enter the task you want to do:")
        if tasks == 's':
            print("\nstopped")
            break
        else:
            tasks_to_do.append(tasks)
        
        
completed_tasks = []
def delete():
    if len(tasks_to_do)>0:
        print("\n enter to stop deleting tasks")
    while len(tasks_to_do) > 0:
        del_task = input("\nenter the task you want to delete: ")
        if del_task == 's':
            print("\nyou have stopped deleting tasks.")
            print("\ndeleted tasks are", completed_tasks)
            print("\nyour remaining tasks are", tasks_to_do)
            break
        elif del_task in tasks_to_do:
            a = tasks_to_do.index(del_task)
            e = tasks_to_do.pop(a)
            completed_tasks.append(e)
            print("\nyour remaining tasks are", tasks_to_do)
        else:
            print("task not in list")
    else:
        print("\n to do list is empty")
        
        
def viewtasks():
    if len(tasks_to_do) > 0:
        print("\n",tasks_to_do)
    else:
        print("\nyour to do list is empty")

        
def del_all():
    tasks_to_do.clear()
    print("your list is empty")

var = True
while var:
    menu()
    task_input = input("Enter your choice: ")
    user_input(task_input)
    
```
